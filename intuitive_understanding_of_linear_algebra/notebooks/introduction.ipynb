```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

a = np.linspace(-5, 5, 30)
b = np.linspace(-5, 5, 30)

indices = np.arange(len(a))
np.random.shuffle(indices)
a = a[indices]
b = b[indices]

a, b = np.meshgrid(a, b)

c = np.random.random(a.shape) * 5
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': '3d'}, figsize=(7, 7))
ax.scatter(a, b, c, c=c, cmap='viridis', label='Data points')
ax.set_proj_type("ortho")

ax.set_xlim(-7, 7)
ax.set_ylim(-7, 7)
ax.set_zlim(-14, 14)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.tick_params(axis='x', labelsize=6)
ax.tick_params(axis='y', labelsize=6)
ax.tick_params(axis='z', labelsize=6)

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': '3d'}, figsize=(7, 7))
scatter = ax.scatter(a, b, c, c=c, cmap='viridis', label='Data points')
ax.set_proj_type("ortho")

ax.plot_wireframe(a, b, np.zeros_like(c) - 7, color='red', alpha=0.3, label='Projected')


ax.set_xlim(-7, 7)
ax.set_ylim(-7, 7)
ax.set_zlim(-14, 14)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.tick_params(axis='x', labelsize=6)
ax.tick_params(axis='y', labelsize=6)
ax.tick_params(axis='z', labelsize=6)
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': '3d'}, figsize=(7, 7))
scatter = ax.scatter(a, b, c, c=c, cmap='viridis', label='Data points')
ax.set_proj_type("ortho")

ax.plot_wireframe(a, b, np.zeros_like(c) - 7, color='red', alpha=0.3, label='Projected')

for i in range(len(a.flatten())):
    ax.plot([a.flatten()[i], a.flatten()[i]], [b.flatten()[i], b.flatten()[i]], zs=[c.flatten()[i], -7], color='red', alpha=0.1)

ax.set_xlim(-7, 7)
ax.set_ylim(-7, 7)
ax.set_zlim(-14, 14)
ax.tick_params(axis='x', labelsize=6)
ax.tick_params(axis='y', labelsize=6)
ax.tick_params(axis='z', labelsize=6)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

ax.legend()

plt.show()
```
