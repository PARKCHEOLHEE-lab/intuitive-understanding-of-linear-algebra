```python
add_path()
from intuitive_understanding_of_linear_algebra.src import utils
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(21, 7))
ax1, ax2, ax3 = axes.flatten()

xlim = (-3, 3)
ylim = (-3, 3)

basis = np.array(
    [
        [1, 0],
        [0, 1],
    ]
).astype(float)

square = np.array(
    [
        [0, 0],
        [1, 0],
        [1, 1],
        [0, 1],
        [0, 0],
    ]
)

utils.plot_full_custom_2d_grid(ax1, basis.T, xlim, ylim)
ax1.set_title("$\\det {A} = ad - bc$ = 1.00 \n")
ax1.fill(square[:, 0], square[:, 1], color="black", label="square", alpha=0.15)
ax1.quiver(0, 0, basis[0][0], basis[1][0], angles="xy", scale_units="xy", scale=1, color="r")
ax1.quiver(0, 0, basis[0][1], basis[1][1], angles="xy", scale_units="xy", scale=1, color="g")
ax1.text(basis[0][0], basis[1][0] - 0.05, f"{list(basis.T[0])}", color="r", fontsize=10, ha="left")
ax1.text(basis[0][1], basis[1][1] + 0.05, f"{list(basis.T[1])}", color="g", fontsize=10, ha="center")
ax1.set_xlim(xlim)
ax1.set_ylim(ylim)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_aspect("equal")
ax1.legend(loc="upper left")

matrix_a = np.array(
    [
        [1, -0.6],
        [0, 0.6]
    ]
)

parallelogram = square @ matrix_a.T

utils.plot_full_custom_2d_grid(ax2, matrix_a.T, xlim, ylim)
ax2.set_title("$\\det A = ad - bc$ = {:.2f} \n".format(np.linalg.det(matrix_a)))
ax2.fill(parallelogram[:, 0], parallelogram[:, 1], color="black", label="parallelogram", alpha=0.15)
ax2.quiver(0, 0, matrix_a[0][0], matrix_a[1][0], angles="xy", scale_units="xy", scale=1, color="r")
ax2.quiver(0, 0, matrix_a[0][1], matrix_a[1][1], angles="xy", scale_units="xy", scale=1, color="g")
ax2.text(matrix_a[0][0], matrix_a[1][0] - 0.05, f"{list(matrix_a.T[0])}", color="r", fontsize=10, ha="left")
ax2.text(matrix_a[0][1], matrix_a[1][1] + 0.05, f"{list(matrix_a.T[1])}", color="g", fontsize=10, ha="center")
ax2.set_xlim(xlim)
ax2.set_ylim(ylim)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_aspect("equal")
ax2.legend(loc="upper left")

scaler = np.array(
    [
        [0.1, 0],
        [0, 0.1],
    ]
)

degree_square = square @ scaler

utils.plot_full_custom_2d_grid(ax3, matrix_a.T, xlim, ylim)
ax3.set_title("$\\det A$ = area of parallelogram = {:.2f} \n".format(np.cross(matrix_a[0], matrix_a[1])))
ax3.fill(parallelogram[:, 0], parallelogram[:, 1], color="black", label="parallelogram", alpha=0.15)
ax3.plot([0, 0], [0, matrix_a[1][1]], linewidth=1, color="black", linestyle="dashed", alpha=0.5)
ax3.plot(degree_square[:, 0], degree_square[:, 1], color="gray", alpha=0.5)
ax3.text(-0.5, 0.25, "height=0.6", ha="right")
ax3.text(0.6, -0.3, "length=1.0", ha="center")
ax3.set_xlim(xlim)
ax3.set_ylim(ylim)
ax3.set_xticks([])
ax3.set_yticks([])
ax3.set_aspect("equal")
ax3.legend(loc="upper left")



plt.show()
```

```python
add_path()
import pyvista
import matplotlib.pyplot as plt
import numpy as np
from intuitive_understanding_of_linear_algebra.src import utils
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

fig = plt.figure(figsize=(14, 14))

ax1 = fig.add_subplot(221, projection="3d")
ax1.set_proj_type("ortho")

xlim = (-2, 2)
ylim = (-2, 2)
zlim = (-2, 2)

basis = np.array(
    [
        [1, 0, 0],
        [0, 1, 0],
        [0, 0, 1],
    ]
).astype(float)

utils.plot_full_custom_3d_grid(ax1, basis.T, xlim, ylim, zlim)

mesh = pyvista.examples.download_cow()

mesh.points *= 0.15
mesh.points[:, [1, 2]] = mesh.points[:, [2, 1]]
mesh.points[:, [0, 1]] = mesh.points[:, [1, 0]]
mesh.points[:, [1]] *= -1
mesh.points[:, [0, 1]] -= 0.5

faces = []
i = 0
while i < len(mesh.faces):
    n = mesh.faces[i]
    faces.append(mesh.faces[i+1:i+1+n])
    i += n + 1

mesh_poly = [[mesh.points[vertex] for vertex in face] for face in faces]

centroid = mesh.points.mean(0)
ax1.scatter(centroid[0], centroid[1], centroid[2], color="purple", label="cow")

ax1.set_title("$\\det A$ = {} \n".format(np.linalg.det(basis)))
ax1.quiver(0, 0, 0, basis[0][0], basis[1][0], basis[2][0], color="r")
ax1.quiver(0, 0, 0, basis[0][1], basis[1][1], basis[2][1], color="g")
ax1.quiver(0, 0, 0, basis[0][2], basis[1][2], basis[2][2], color="b")
ax1.add_collection3d(Poly3DCollection(mesh_poly, linewidths=0.5, color="purple", edgecolors="purple", alpha=0.5))
ax1.set_xlim(xlim)
ax1.set_ylim(ylim)
ax1.set_zlim(zlim)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_zticks([])
ax1.set_aspect("auto")
ax1.legend(loc="upper left")

ax2 = fig.add_subplot(222, projection="3d")
ax2.set_proj_type("ortho")

matrix_a = np.array(
    [
        [1, 0, 0],
        [0, 1.3, 0],
        [0, 0, -1],
    ]
)

reflected_mesh = mesh.copy()
reflected_mesh.points = (matrix_a @ reflected_mesh.points.T).T 
reflected_mesh_poly = [[reflected_mesh.points[vertex] for vertex in face] for face in faces]

centroid = reflected_mesh.points.mean(0)
ax2.scatter(centroid[0], centroid[1], centroid[2], color="purple", label="cow")

utils.plot_full_custom_3d_grid(ax2, matrix_a.T, xlim, ylim, zlim)
ax2.set_title("$\\det B$ = {} \n".format(np.linalg.det(matrix_a)))
ax2.quiver(0, 0, 0, matrix_a[0][0], matrix_a[1][0], matrix_a[2][0], color="r")
ax2.quiver(0, 0, 0, matrix_a[0][1], matrix_a[1][1], matrix_a[2][1], color="g")
ax2.quiver(0, 0, 0, matrix_a[0][2], matrix_a[1][2], matrix_a[2][2], color="b")
ax2.add_collection3d(Poly3DCollection(reflected_mesh_poly, linewidths=0.5, color="purple", edgecolors="purple", alpha=0.5))
ax2.set_xlim(xlim)
ax2.set_ylim(ylim)
ax2.set_zlim(zlim)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_zticks([])
ax2.set_aspect("auto")
ax2.legend(loc="upper left")

ax3 = fig.add_subplot(223, projection="3d")
ax3.set_proj_type("ortho")

matrix_a = np.array(
    [
        [-1, 0, 0],
        [0, 0.6, 0],
        [0, 0, 1.2],
    ]
)

reflected_mesh = mesh.copy()
reflected_mesh.points = (matrix_a @ reflected_mesh.points.T).T 
reflected_mesh_poly = [[reflected_mesh.points[vertex] for vertex in face] for face in faces]

centroid = reflected_mesh.points.mean(0)
ax3.scatter(centroid[0], centroid[1], centroid[2], color="purple", label="cow")

utils.plot_full_custom_3d_grid(ax3, matrix_a.T, xlim, ylim, zlim)
ax3.set_title("$\\det C$ = {} \n".format(np.linalg.det(matrix_a)))
ax3.quiver(0, 0, 0, matrix_a[0][0], matrix_a[1][0], matrix_a[2][0], color="r")
ax3.quiver(0, 0, 0, matrix_a[0][1], matrix_a[1][1], matrix_a[2][1], color="g")
ax3.quiver(0, 0, 0, matrix_a[0][2], matrix_a[1][2], matrix_a[2][2], color="b")
ax3.add_collection3d(Poly3DCollection(reflected_mesh_poly, linewidths=0.5, color="purple", edgecolors="purple", alpha=0.5))
ax3.set_xlim(xlim)
ax3.set_ylim(ylim)
ax3.set_zlim(zlim)
ax3.set_xticks([])
ax3.set_yticks([])
ax3.set_zticks([])
ax3.set_aspect("auto")
ax3.legend(loc="upper left")

ax4 = fig.add_subplot(224, projection="3d")
ax4.set_proj_type("ortho")

matrix_a = np.array(
    [
        [-1.5, 0, 0],
        [0, -0.7, 0],
        [0, 0, 1],
    ]
)

reflected_mesh = mesh.copy()
reflected_mesh.points = (matrix_a @ reflected_mesh.points.T).T 
reflected_mesh_poly = [[reflected_mesh.points[vertex] for vertex in face] for face in faces]

centroid = reflected_mesh.points.mean(0)
ax4.scatter(centroid[0], centroid[1], centroid[2], color="purple", label="cow")

utils.plot_full_custom_3d_grid(ax4, matrix_a.T, xlim, ylim, zlim)
ax4.set_title("$\\det D$ = {:.2f} \n".format(np.linalg.det(matrix_a)))
ax4.quiver(0, 0, 0, matrix_a[0][0], matrix_a[1][0], matrix_a[2][0], color="r")
ax4.quiver(0, 0, 0, matrix_a[0][1], matrix_a[1][1], matrix_a[2][1], color="g")
ax4.quiver(0, 0, 0, matrix_a[0][2], matrix_a[1][2], matrix_a[2][2], color="b")
ax4.add_collection3d(Poly3DCollection(reflected_mesh_poly, linewidths=0.5, color="purple", edgecolors="purple", alpha=0.5))
ax4.set_xlim(xlim)
ax4.set_ylim(ylim)
ax4.set_zlim(zlim)
ax4.set_xticks([])
ax4.set_yticks([])
ax4.set_zticks([])
ax4.set_aspect("auto")
ax4.legend(loc="upper left")

plt.show()
```
