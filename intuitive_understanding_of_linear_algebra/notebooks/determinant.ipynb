```python
add_path()
from intuitive_understanding_of_linear_algebra.src import utils
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(14, 7))
ax1, ax2 = axes.flatten()

xlim = (-3, 3)
ylim = (-3, 3)

basis = np.array(
    [
        [1, 0],
        [0, 1],
    ]
)

square = np.array(
    [
        [0, 0],
        [1, 0],
        [1, 1],
        [0, 1],
        [0, 0],
    ]
)

utils.plot_full_custom_2d_grid(ax1, basis.T, xlim, ylim)
ax1.set_title("$\\det {A} = ad - bc$ = 1.0 \n")
ax1.fill(square[:, 0], square[:, 1], color="black", label="square", alpha=0.15)
ax1.quiver(0, 0, basis[0][0], basis[1][0], angles="xy", scale_units="xy", scale=1, color="r")
ax1.quiver(0, 0, basis[0][1], basis[1][1], angles="xy", scale_units="xy", scale=1, color="g")
ax1.text(basis[0][0], basis[1][0] - 0.05, f"{basis.T[0]}", color="r", fontsize=10, ha="left")
ax1.text(basis[0][1], basis[1][1] + 0.05, f"{basis.T[1]}", color="g", fontsize=10, ha="center")
ax1.set_xlim(xlim)
ax1.set_ylim(ylim)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_aspect("equal")
ax1.legend(loc="upper left")

matrix_a = np.array(
    [
        [1, -0.3],
        [-0.7, 0.6]
    ]
)

parallelogram = square @ matrix_a.T

utils.plot_full_custom_2d_grid(ax2, matrix_a.T, xlim, ylim)
ax2.set_title("$\\det {A} = ad - bc$ = 0.39 \n")
ax2.fill(parallelogram[:, 0], parallelogram[:, 1], color="black", label="parallelogram", alpha=0.15)
ax2.quiver(0, 0, matrix_a[0][0], matrix_a[1][0], angles="xy", scale_units="xy", scale=1, color="r")
ax2.quiver(0, 0, matrix_a[0][1], matrix_a[1][1], angles="xy", scale_units="xy", scale=1, color="g")
ax2.text(matrix_a[0][0], matrix_a[1][0] - 0.05, f"{list(matrix_a.T[0])}", color="r", fontsize=10, ha="left")
ax2.text(matrix_a[0][1], matrix_a[1][1] + 0.05, f"{list(matrix_a.T[1])}", color="g", fontsize=10, ha="center")
ax2.set_xlim(xlim)
ax2.set_ylim(ylim)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_aspect("equal")
ax2.legend(loc="upper left")

plt.show()
```
