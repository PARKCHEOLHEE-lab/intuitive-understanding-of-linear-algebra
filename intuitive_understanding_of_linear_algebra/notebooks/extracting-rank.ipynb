```python
add_path()
import matplotlib.pyplot as plt
import numpy as np
from intuitive_understanding_of_linear_algebra.src import utils
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

fig = plt.figure(figsize=(21, 7))

xlim = (-2, 2)
ylim = (-2, 2)
zlim = (-2, 2)

ax1 = fig.add_subplot(131, projection="3d")
ax1.set_proj_type("ortho")

a1 = -np.array([[0.5], [0.5], [0.5]])
a2 = a1 * -1.5
a3 = a1 * 3
a = np.hstack([a1, a2, a3])

utils.plot_full_custom_3d_grid(ax1, a.T, xlim, ylim, zlim, linewidth=0.1, alpha=0.1, color="black")

ax1.set_title("span{$a_1, a_2, a_3$}" + f"\n rank $A$ = {np.linalg.matrix_rank(a)}")
ax1.quiver(0, 0, 0, a[0][0], a[1][0], a[2][0], color="r", arrow_length_ratio=0.33, label="$a_1$")
ax1.quiver(0, 0, 0, a[0][1], a[1][1], a[2][1], color="g", arrow_length_ratio=0.22, label="$a_2$")
ax1.quiver(0, 0, 0, a[0][2], a[1][2], a[2][2], color="b", arrow_length_ratio=0.11, label="$a_3$")
ax1.set_xlim(xlim)
ax1.set_ylim(ylim)
ax1.set_zlim(zlim)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_zticks([])
ax1.set_aspect("auto")
ax1.legend(loc="upper left")


ax2 = fig.add_subplot(132, projection="3d")
ax2.set_proj_type("ortho")

b1 = np.array([[0.5], [1.3], [1]])
b2 = np.array([[-1], [-0.8], [0]])
b3 = b1 * -0.7 + b2
b4 = b2 * -1.2 + b1 * -1
b = np.hstack([b1, b2, b3, b4])

ax2.set_title("span{$b_1, b_2, b_3, b_4$}" + f"\n rank $B$ = {np.linalg.matrix_rank(b)}")
ax2.quiver(0, 0, 0, b[0][0], b[1][0], b[2][0], color="r", arrow_length_ratio=0.2, label="$b_1$")
ax2.quiver(0, 0, 0, b[0][1], b[1][1], b[2][1], color="g", arrow_length_ratio=0.2, label="$b_2$")
ax2.quiver(0, 0, 0, b[0][2], b[1][2], b[2][2], color="b", arrow_length_ratio=0.13, label="$b_3$")
ax2.quiver(0, 0, 0, b4[0][0], b4[1][0], b4[2][0], color="magenta", arrow_length_ratio=0.3, label="$b_4$")
utils.plot_full_custom_3d_grid(ax2, b[:, :3].T, xlim, ylim, zlim, linewidth=0.2, grid_step=1, color="black")

ax2.set_xlim(xlim)
ax2.set_ylim(ylim)
ax2.set_zlim(zlim)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_zticks([])
ax2.set_aspect("auto")
ax2.legend(loc="upper left")


ax3 = fig.add_subplot(133, projection="3d")
ax3.set_proj_type("ortho")

c1 = np.array([[0.5], [1], [0]])
c2 = np.array([[1.5], [-1.7], [0]])
c3 = np.array([[0.7], [0.8], [1.1]])
c4 = -c1 * 1.5
c5 = -c2 + c4
c = np.hstack([c1, c2, c3, c4, c5])

ax3.set_title("span{$c_1, c_2, c_3, c_4, c_5$}" + f"\n rank $C$ = {np.linalg.matrix_rank(c)}")
ax3.quiver(0, 0, 0, c[0][0], c[1][0], c[2][0], color="r", arrow_length_ratio=0.22, label="$c_1$")
ax3.quiver(0, 0, 0, c[0][1], c[1][1], c[2][1], color="g", arrow_length_ratio=0.2, label="$c_2$")
ax3.quiver(0, 0, 0, c[0][2], c[1][2], c[2][2], color="b", arrow_length_ratio=0.15, label="$c_3$")
ax3.quiver(0, 0, 0, c4[0][0], c4[1][0], c4[2][0], color="magenta", arrow_length_ratio=0.15, label="$c_4$")
ax3.quiver(0, 0, 0, c5[0][0], c5[1][0], c5[2][0], color="c", arrow_length_ratio=0.1, label="$c_5$")
utils.plot_full_custom_3d_grid(ax3, c[:, :3].T, xlim, ylim, zlim, linewidth=0.3, grid_step=1, color="black")

ax3.set_xlim(xlim)
ax3.set_ylim(ylim)
ax3.set_zlim(zlim)
ax3.set_xticks([])
ax3.set_yticks([])
ax3.set_zticks([])
ax3.set_aspect("auto")
ax3.legend(loc="upper left")

plt.show()
```
