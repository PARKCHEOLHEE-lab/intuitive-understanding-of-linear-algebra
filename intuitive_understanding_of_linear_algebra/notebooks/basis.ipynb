```python
import matplotlib.pyplot as plt
import numpy as np


# Define the standard basis vectors and the vector v
e1 = np.array([1, 0])
e2 = np.array([0, 1])

e1_scaler = 3
e2_scaler = 2
e1_3 = e1 * e1_scaler
e2_2 = e2 * e2_scaler
v = e1_3 + e2_2

fig, ax = plt.subplots(figsize=(7, 7))
ax.set_aspect('equal')

# Plot the basis vectors
ax.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='r', label='$\\vec{e}_1$')
ax.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='g', label='$\\vec{e}_2$')

# Plot the scaled vectors for visualization
ax.quiver(0, 0, e1_3[0], e1_3[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
ax.quiver(e1_3[0], e1_3[1], e2_2[0], e2_2[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)

# Plot the vector v
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='black', label='$\\vec{v_1}$')

# Set the limits of the plot
xlim = (-2, 5)
ylim = (-3, 4)
ax.set_xlim(xlim)
ax.set_ylim(ylim)

# Plot 2d grid using basis
basis = np.array([e1, e2])
utils.plot_full_custom_2d_grid(ax, basis, xlim, ylim)

ax.set_xticks([])
ax.set_yticks([])

ax.set_title('Visualization of Standard Basis Vectors $\\vec{e}_1$ and $\\vec{e}_2$ and Vector $\\vec{v_1}$ \n')
ax.text(e1[0], e1[1], '$\\vec{e}_1$', color='r', fontsize=12, ha='left')
ax.text(e2[0], e2[1], '$\\vec{e}_2$', color='g', fontsize=12, ha='center')
ax.text(v[0], v[1], '$\\vec{v_1}$', color='black', fontsize=12, ha='center')
ax.text(-0.1, -1.7, f'$\\vec{{v_1}}$ = {e1_scaler}$\\vec{{e}}_1$ + {e2_scaler}$\\vec{{e}}_2$', color='black', fontsize=25)
ax.legend()

plt.show()
```
