```python
add_path()
from intuitive_understanding_of_linear_algebra.src import utils
import matplotlib.pyplot as plt
import numpy as np


# Define the standard basis vectors and the vector v
e1 = np.array([1, 0])
e2 = np.array([0, 1])

e1_scaler = 3
e2_scaler = 2
e1_3 = e1 * e1_scaler
e2_2 = e2 * e2_scaler
v = e1_3 + e2_2

fig, ax = plt.subplots(figsize=(7, 7))
ax.set_aspect("equal")

# Plot the basis vectors
ax.quiver(0, 0, e1[0], e1[1], angles="xy", scale_units="xy", scale=1, color="r", label="$\\vec{e}_1$")
ax.quiver(0, 0, e2[0], e2[1], angles="xy", scale_units="xy", scale=1, color="g", label="$\\vec{e}_2$")

# Plot the scaled vectors for visualization
ax.quiver(0, 0, e1_3[0], e1_3[1], angles="xy", scale_units="xy", scale=1, color="gray", alpha=0.5)
ax.quiver(e1_3[0], e1_3[1], e2_2[0], e2_2[1], angles="xy", scale_units="xy", scale=1, color="gray", alpha=0.5)

# Plot the vector v
ax.quiver(0, 0, v[0], v[1], angles="xy", scale_units="xy", scale=1, color="black", label="$\\vec{v_1}$")

# Set the limits of the plot
xlim = (-2, 5)
ylim = (-3, 4)
ax.set_xlim(xlim)
ax.set_ylim(ylim)

# Plot 2d grid using basis
basis = np.array([e1, e2])
utils.plot_full_custom_2d_grid(ax, basis, xlim, ylim)

ax.set_xticks([])
ax.set_yticks([])

ax.set_title("Visualization of Standard Basis Vectors $\\vec{e}_1$ and $\\vec{e}_2$ and Vector $\\vec{v_1}$ \n")
ax.text(e1[0], e1[1], "$\\vec{e}_1$", color="r", fontsize=12, ha="left")
ax.text(e2[0], e2[1], "$\\vec{e}_2$", color="g", fontsize=12, ha="center")
ax.text(v[0], v[1], "$\\vec{v_1}$", color="black", fontsize=12, ha="center")
ax.text(-0.1, -1.7, f"$\\vec{{v_1}}$ = {e1_scaler}$\\vec{{e}}_1$ + {e2_scaler}$\\vec{{e}}_2$", color="black", fontsize=25)
ax.legend()

plt.show()
```

```python
add_path()
from intuitive_understanding_of_linear_algebra.src import utils
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 4, figsize=(28, 7))
ax1, ax2, ax3, ax4 = axes.flatten()

xlim = (-5, 5)
ylim = (-5, 5)

e1 = np.array([1, 0.2])
e2 = np.array([0.2, 1.5])
utils.plot_full_custom_2d_grid(ax1, np.array([e1, e2]), xlim, ylim)
ax1.set_title("[✔️] Both conditions for becoming basis are satisfied \n")
ax1.quiver(0, 0, e1[0], e1[1], angles="xy", scale_units="xy", scale=1, color="r", label="$\\vec{e}_1$")
ax1.quiver(0, 0, e2[0], e2[1], angles="xy", scale_units="xy", scale=1, color="g", label="$\\vec{e}_2$")
ax1.set_xlim(-5, 5)
ax1.set_ylim(-5, 5)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_aspect("equal")
ax1.legend(loc="upper left")

e1 = np.array([1, 0])
e2 = np.array([0, 1])
e3 = np.array([2, 2])
utils.plot_full_custom_2d_grid(ax2, np.array([e1, e2]), xlim, ylim)
utils.plot_full_custom_2d_grid(ax2, np.array([e2, e3]), xlim, ylim)
ax2.set_title("[✖️] $\\vec{e_3}$ is not linearly independent \n")
ax2.quiver(0, 0, e1[0], e1[1], angles="xy", scale_units="xy", scale=1, color="r", label="$\\vec{e}_1$")
ax2.quiver(0, 0, e2[0], e2[1], angles="xy", scale_units="xy", scale=1, color="g", label="$\\vec{e}_2$")
ax2.quiver(0, 0, e3[0], e3[1], angles="xy", scale_units="xy", scale=1, color="b", label="$\\vec{e}_3$")
ax2.set_xlim(-5, 5)
ax2.set_ylim(-5, 5)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_aspect("equal")
ax2.legend(loc="upper left")

e1 = np.array([0.5, -0.7])
utils.plot_full_custom_2d_grid(ax3, np.array([e1, e1]), xlim, ylim)
ax3.set_title("[✖️] $\\vec{e_1}$ cannot span all $\\mathbb{R}^2$ space \n")
ax3.quiver(0, 0, e1[0], e1[1], angles="xy", scale_units="xy", scale=1, color="r", label="$\\vec{e}_1$")
ax3.set_xlim(-5, 5)
ax3.set_ylim(-5, 5)
ax3.set_xticks([])
ax3.set_yticks([])
ax3.set_aspect("equal")
ax3.legend(loc="upper left")

e1 = np.array([2, 2])
e2 = np.array([1, 1])
utils.plot_full_custom_2d_grid(ax4, [e1, e2], xlim, ylim)
ax4.set_title("[✖️] $\\vec{e_1}$ and $\\vec{e_2}$ are not linearly independent of each other \n")
ax4.quiver(0, 0, e1[0], e1[1], angles="xy", scale_units="xy", scale=1, color="r", label="$\\vec{e}_1$")
ax4.quiver(0, 0, e2[0], e2[1], angles="xy", scale_units="xy", scale=1, color="g", label="$\\vec{e}_2$")
ax4.set_xlim(-5, 5)
ax4.set_ylim(-5, 5)
ax4.set_xticks([])
ax4.set_yticks([])
ax4.set_aspect("equal")
ax4.legend(loc="upper left")

plt.show()
```
