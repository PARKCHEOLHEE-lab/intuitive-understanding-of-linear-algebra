```python
import matplotlib.pyplot as plt
import numpy as np

# Define the basis vectors and the vector v
e1 = np.array([1, 0])
e2 = np.array([0, 1])
v = np.array([3, 2])

e1_3 = e1 * 3
e2_2 = e2 * 2

# Plot the basis vectors
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='r', label='$\\vec{e}_1$')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='g', label='$\\vec{e}_2$')

plt.quiver(0, 0, e1_3[0], e1_3[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
plt.quiver(e1_3[0], e1_3[1], e2_2[0], e2_2[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)

# Plot the vector v
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='black', label='$\\vec{v}$')

# Set the limits of the plot
plt.xlim(-2, 5)
plt.ylim(-3, 4)

# Add grid, legend and labels
# plt.grid()
plt.legend()
plt.title('Visualization of Basis Vectors $\\vec{e}_1$ and $\\vec{e}_2$ and Vector $\\vec{v}$ \n')

plt.text(e1[0], e1[1], '$\\vec{e}_1$', color='r', fontsize=12, ha='left')
plt.text(e2[0], e2[1], '$\\vec{e}_2$', color='g', fontsize=12, ha='center')
plt.text(v[0], v[1], '$\\vec{v}$', color='black', fontsize=12, ha='center')
plt.text(0.5, -1.7, '$\\vec{v}$ = 3$\\vec{e}_1$ + 2$\\vec{e}_2$', color='black', fontsize=15)

# Show the plot
plt.show()
```
